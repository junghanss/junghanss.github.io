## NumPy

NumPy (Numerical Python) is an open source Python library that’s used in almost every field of science and engineering. It’s the universal standard for working with numerical data in Python, and it’s at the core of the scientific Python and PyData ecosystems.

```python
import numpy as np
```

```python
# Creación de arrays

array = np.array([1, 2, 3, 4, 5, 6])  # Vector de 1x6
print(array)

# Array en 2D:
matrix = np.array([[1,2],[3,4]])   # Matriz de 2x2
print(matrix)

# Array de ceros:
print(np.zeros(8))

# Array de unos:
print(np.ones(6))

# Array con numeros de un intervalo (primeros dos argumentos) y un step especifico entre estos (3er argumento)
print(np.arange(2, 11, 2))

# Array con una cantidad especifica de numeros (num=5) de un intervalo (primeros dos argumentos):
print(np.linspace(0, 10, num=5))
```

```python
# Ordenamiento de los valores del array:
a = np.array([5,0,3,65,10])
print(np.sort(a))

# Concatenamiento de arrays:
a2 = np.concatenate((a, np.ones(5)))
print(a2)

x = np.array([[1, 2], [3, 4]])
y = np.array([[5, 6]])
z = np.concatenate((x,y), axis=0)
print(z)
```

```python
# Podemos "apilarlos" (stack) sea horizontal o verticalmente:

print(np.vstack((x, y)))

print(np.hstack((np.ones(2), np.zeros(2))))
```

*Some definitions:*

- ndarray.ndim will tell you the number of axes, or dimensions, of the array.

- ndarray.size will tell you the total number of elements of the array. This is the product of the elements of the array’s shape.

- ndarray.shape will display a tuple of integers that indicate the number of elements stored along each dimension of the array. If, for example, you have a 2-D array with 2 rows and 3 columns, the shape of your array is (2, 3).

```python
array_example = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],

                          [[0, 1, 2, 3],
                           [4, 5, 6, 7]],

                          [[0 ,1 ,2, 3],
                           [4, 5, 6, 7]]])
```

```python
array_example.ndim
```

```python
array_example.size
```

```python
array_example.shape
```

```python
# Podemos cambiar las dimensiones del array con el método reshape

np.reshape(array_example, newshape=(2,3,4), order='C')
```

```python
x = np.arange(1, 25).reshape(2, 12)
print(x)
```

```python
# También podemos transponer matrices de dos formas:

arr = np.arange(6).reshape((2, 3))
print(arr)

# (1) Usando el método np.transpose()
print(arr.transpose())

# (2) Escribiendo el nombre del objeto seguido de ".T"
print(arr.T)
```

#### Operaciones matemáticas con NumPy

- Funciones matemáticas: https://numpy.org/doc/stable/reference/routines.math.html

- Algebra: https://numpy.org/doc/stable/reference/routines.linalg.html

```python
a = np.array([10,20,100,200,500])
b = np.array([3,4,5,6,7])

# Suma:
# (1) Con el método np.add()
print(np.add(a, b))
# (2) simplemente sumando los objetos
print(a+b)

# Lo mismo aplica para restas con np.substract(), divisiones con np.divide()

# Multiplicaciones: se usa simplemente el operador *

print(a*b)

# Power: para elevar a un exponente una base
print(np.power(a,b))
```

```python
# Print the square root of 2
print(np.sqrt(2))

# Print the factorial of 5
print(np.math.factorial(5))

# Print the mathematical constants pi and e
print(np.math.pi)
print(np.math.e)
```

```python
# Producto escalar de dos arrays:
print(np.dot(3, 4))

print(np.dot([2j, 3j], [2j, 3j])) # Inclusive computa operaciones con números complejos

a = [[1, 0], [0, 1]]
b = [[4, 1], [2, 2]]

print(np.dot(a,b))
```

```python
# Producto matricial de dos arrays:
a = np.array([[5, 0],
              [0, 5]])

b = np.array([[1, 2],
             [3,4]])

# (1) Con el método matmul
print(np.matmul(a,b))

# (2) Con el operador @
print(a@b)
```

## Matplotlib


Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()  # Se crea un objeto para la figura y ejes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Se grafica la data dentro
```

```python
# Create an array of x values from -6 to 6
x = np.arange(-6,6,0.001)

# Create a variable y equal to the sine of x
y = np.sin(x)

# Use the plot function to plot x on the horizontal axis and y on the vertical axis
plt.plot(x,y)

# Add a title and axis labels
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Create an array of x values from 0.01 to 5
x = np.arange(0.01,5,0.011)

# Create y variables
y1 = np.log(x)
y2 = 1/x

# Use the plot function to plot the lines
plt.plot(x,y1,'b-',linewidth=3,alpha = 0.6,label='$\log(x)$')
plt.plot(x,y2,'m--',linewidth=3,alpha = 0.6,label='$1/x$')

# Add a title and axis labels
plt.title('Two functions')
plt.xlabel('x')
plt.ylabel('y')

# Set axis limits
plt.xlim([0,5])
plt.ylim([-2,4])

# Make legend
plt.legend(loc='lower right',ncol=2)

# Add grid
plt.grid()
```

```python
# Create an array of x values from -6 to 6
x = np.arange(-6,6,0.001)

# Create y variables
y = np.sin(x)

# Create a new figure that is twice as wide as the default
fig = plt.figure(figsize=(12,4))

# Create axis within the figure
ax1 = fig.add_subplot(1,1,1)

# Plot
ax1.plot(x,y,lw=3,alpha = 0.6)

# Add grid
ax1.grid()
```

```python
# Data generada random con NumPy

np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

```

```python
# Creacion de un grafico tipo scatterplot

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter('a', 'b', c='c', s='d', data=data)

# Seteo de labels
ax.set_xlabel('entry a') 
ax.set_ylabel('entry b') 
```
